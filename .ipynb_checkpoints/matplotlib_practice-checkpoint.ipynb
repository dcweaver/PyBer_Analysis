```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker = "*", color = "blue", linewidth = 2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)

#set x-limit to be january - june
#plt.xlim(0,5)
#plt.xlim("Jan", "June")

# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
#add a grid
plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create Bar plot
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color = "magenta", label = "Boston")
plt.gca().invert_yaxis() #GCA = get current axes (needed to invert axis)
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend() #NEEDED TO DISPLAY LABEL
```

# SCATTERPLOTS

```python
plt.plot(x_axis, y_axis, "o")
```

```python
#SKILL DRILL
plt.scatter(y_axis, x_axis, color = "red", label = "Chicago")
plt.gca().invert_yaxis()
plt.xlabel("Month")
plt.ylabel("Fare($)")
plt.xlim(0,45)
#plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s=y_axis) #adjust bubble size based on y axis
```

```python
y_axis_larger = [] #initialize empty list
for data in y_axis: 
  y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# LIST COMPREHENSION SYNTAX
# new_list = [expression for item in list if conditional]
```

```python
plt.scatter(x_axis, y_axis, s=[blerg * 3 for blerg in y_axis])
```

# Object oriented scatterplot

```python
fig, ax = plt.subplots()
ax.scatter(x_axis,y_axis)
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= [i*5 for i in y_axis], color = "skyblue",
           alpha = 0.8, edgecolors ="black", linewidth = 2, label = "Boston")
ax.legend()
ax.set_title("PyBer fare by Month")


```
